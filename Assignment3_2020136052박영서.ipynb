* 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

* 2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

```python
def addall(list): # addall 함수(for ~ in 리터럴)
    sum = 0
    for i in list :
        sum += i
    return sum

def addallodd(list): # addallodd 함수(for ~ in 리터럴)
    sum = 0
    for i in list :
        if i % 2 == 1 :
            sum += i
    return sum

#=========#

print(addall([1]))
print(addall([1, 2, 3, 4, 5, 6, 7, 8, 9]))
print(addallodd([1]))
print(addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

```python
def addall(list): # addall 함수(리스트 내포)
    return sum([i for i in list])

def addallodd(list): # addallodd 함수(리스트 내포)
    return sum([i for i in list if i % 2 == 1])

#=========#

print(addall([1]))
print(addall([1, 2, 3, 4, 5, 6, 7, 8, 9]))
print(addallodd([1]))
print(addallodd([1, 2, 3, 4, 5, 6, 7, 8, 9]))
```

    다음 문제를 정확히 읽고 적합한 코딩을 제시하시오.
* length(X)를 다음과 같이 정의한다 : 10진수로 표현된 X의 자릿수
* Strange라는 숫자를 아래과 같이 정의한다.
    * 음이 아닌 정수 중 자릿수가 1인 모든 숫자는 Strange이다.
    * length(X) > 1 인 숫자 X 는 다음 2가지를 만족하면 Strange이다.
        * X는 length(X)로 정확히 나누어 떨어진다.
        * X / length(X) 가 strnage이다.
* 이 문제에서 숫자 하나가 주어질 때, 이 숫자가 Strange인지 판단하는 함수 isStrange(num)을 작성하시오.
* isStrange(num) 함수가 재귀적인 함수가 되도록 작성해 보세요.

```python
def length(X) :
    se = 1
    digit = 1

    while (1) :
        if (X/se < 10) :
            return(digit)
            break
        digit += 1
        se = str(se)
        se += '0'
        se = int(se)
    return digit

#========#

def isStrange(num) :
    if length(num) == 1 :
        return True
    else :
        if num % length(num) == 0 and isStrange(num / length(num)) == True :
            return True

    return False


print(isStrange(5))
print(isStrange(10))
print(isStrange(100))
print(isStrange(7882602506492313600))
print(isStrange(6012302549522098651))
```

> length()함수는 자릿수를 리턴하는 함수. Strange() 함수는 length()에서 자리수를 받아와 만약 자리수가 1인 경우에는 True를 리턴하고, X가 leng(x)로 정확히 나누어떨어지면서 X / length(X) 가 strange였을때 True를 리턴한다, 상기의 경우들과 다른 경우에는 False를 리턴한다.

```python
def solve(a) :
    M = sum(a)
    m = max(a)
    m_ = min(a)
    ans = []

    if m_ == m :
        for i in range(1, len(a)+1):
            if len(a) % i == 0:
                ans.append(i*a[0])
    else :    
        for i in range(m, M+1) :
            if (M % i == 0) :
                ans.append(i)

    return ans

print(solve([1, 2, 3]))
print(solve([1, 2, 1, 1, 1, 2, 1, 3]))
print(solve([1, 1, 1, 1]))
print(solve([9999] * 10**5 ))
```

> M은 리스트의 총합, m_은 리스트 내 원소들의 최소값, m은 리스트 내 원소들의 최고값이다. 만약 m_ == m일때(즉 모든 리스트의 원소들의 값이 똑같을 경우에는) 리스트의 길이를 계산하여 리스트의 길이의 약수와 원소들을 곱하여 구한다.

    이전 Assignment 2의 마지막 문제는 웹 URL로 지정된 웹페이지를 문자열로 가져와 모든 HTML 태그를 제외한 순수 텍스트를 얻어내고 그 안에 존재하는 단어를 추출하고 단어들의 총 개수를 출력하는 프로그램을 작성하는 것이었다. 이번에는 그 마지막 숙제를 그대로 확장하여 웹 URL로 지정된 웹페이지 내 순수 텍스트 안에 존재하는 각 단어들에 대해 다음 요구사항 대로 출력하는 프로그램을 작성하시오

```python
import re
import requests
req = requests.get('http://www.naver.com', verify=False)
source = req.text


a = source.split("\n") # 줄내림 제거
listb = []
final = []

for i in range(0, len(a)-1) :
    listb.append(a[i].strip()) # 요소 앞뒤의 공백문자 제거


for i in range(1, len(listb)-1) :
    if (listb[i].startswith(('<!', '//-->', '</'))) :
        continue
    if (listb[i].startswith(('<html', '<script', '<head', '<meta', '<title', '</', '<form'))) :
        continue
    if (listb[i].startswith(('<input'))) :
        continue
    if (listb[i].startswith(('<body', '<span', '<li', '<div', '<ul', '<aside', '<a href='))) :
        continue
    if (listb[i].startswith(('<!'))) :
        continue
    final.append(listb[i]) 
    
    # 일차적으로 html 태그를 줄임
    
final2=[]

for i in range(0, len(final)-1) :
    
    final[i] = re.sub('</?[\w*].>', '',final[i])
    final[i] = re.sub('<.*=".*">', '',final[i])     
    final[i] = re.sub('</?.*/?>', '',final[i])    
    final[i] = re.sub('<p id=".*">', '',final[i])     

    # 남은 html 태그들을 정규식을 이용해 제거
    
final2 = []    
    
for i in range(0, len(final)-1) :
    if final[i] == '' :
        continue
    final2.append(final[i])

    # 리스트 내의 공백 요소들을 제거
    
listA = []
output = []   

for i in range(0, len(final2)-1) :
    listA.append(final2[i].split())

    # listA에 공백 문자로 나누어진 요소를 넣음
    
for A in listA :
    output += A
    
    # 중첩된 리스트를 풀고 일차원 리스트로 전환
 
fre = dict()

# 빈도 출력

for i in output:
    fre[i] = 0
for i in fre.keys():
    for j in output:
        if ( j == i):
            fre[i]+=1
print(fre)

# print(output)
print("해당 웹페이지의 단어 개수는 ", len(output), "개 입니다.", sep="")
```
